```python
plt.figure() # making a figure on the page ( ther reactangle)
plt.subplot(1, 2, 1)# on one page make two sub parts and access the first sub part
plt.plot(x, np.sin(x))

plt.subplot(1,2,2) # second part of plot
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.figure() # making a figure on the page ( their reactangle)
plt.subplot(2, 2, 1)# on one page make two sub parts and access the first sub part
plt.plot(x, np.sin(x))

plt.subplot(2,2,2) # second part of plot
plt.plot(x,np.cos(x))

plt.subplot(2,2,3) # second part of plot
plt.plot(x,np.tan(x))

plt.subplot(2,2,4) # second part of plot
plt.plot(x,np.cos(x +1))
plt.show()
```

```python
plt.figure() # making a figure on the page ( ther reactangle)
plt.subplot(2, 1, 1)# on one page make two sub parts and access the first sub part
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)# on one page make two sub parts and access the first sub part
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.plot(x, np.sin(x-1), color = 'green')
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')  # use the entire plot
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid() # get it on number system
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid() # get it on number system
plt.title("this is sin plot before real dataset execution to udnerstand basic syntax")
plt.axis('equal')
plt.show()
```

```python
# provide x and y asix names
plt.plot(x, np.sin(x-0), color = 'blue', label = 'Sinx')
plt.plot(x, np.cos(x-0), color = 'red', label = 'Cosx')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid() # get it on number system
plt.xlabel("thi sis x axis")
plt.ylabel("this is called y axis")
# plt.axis('equal')
plt.legend() #units
plt.show()
```

```python
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
# use this method to visulaize in 3 d
ax = fig.add_subplot(projection = '3d')
#------------------------------------------
ax.scatter(x, y, z, c = 'blue', alpha = .9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
